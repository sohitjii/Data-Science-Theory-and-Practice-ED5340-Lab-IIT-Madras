5.  Text File Input Output

Create a .txt (text) file and use the pledge of India as the content of the text file.

Write a python program that reads this text file, processes it by counting the number of occurrences of each word in the file, and then writes the result back to a new text file.

```python
def create_dic(f): 
    with open('pledge.txt','r') as f:
        content=f.read()
    print(content)


    from collections import defaultdict
    dict1=defaultdict()
    text=""
    for i in range(len(content)):
        if content[i]=="." or content[i]==" ":
                if text.lower() in dict1:
                    dict1[text.lower()]+=1
                else:
                    dict1[text.lower()]=1
                text=""
        text+=content[i]
    print(dict1)
```

```python
f=open('pledge.txt','r')
create_dic(f)

f = open("pledge_output.txt","w")
for key,val in dict1.items():
    string = str(key) +" : "+ str(val)+"\n"
    
    f.write(string)
f.close()
```
